```python
import seaborn as sns
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('ggplot')
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
```

```python
df=pd.read_csv("movies.csv")
```

```python
df.head()
```

```python
# Shape of data
df.shape
```

```python
#Columns in the data frame
df.columns
```

```python
# Checking for null values in data

for col in df.columns:
    print(col," - ",np.sum(df[col].isnull()))
```

```python
#Handling null values
df.dropna(inplace=True)
```

```python
#Datatypes of each column
df.dtypes
```

```python
#Changing Data types to int for Votes, Budget and Gross
df['votes']=df['votes'].astype('int64')
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
```

```python
#Sorting movies based on gross collections
df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
#Dropping duplicate rows
df=df.drop_duplicates()
```

Find Correlation between columns

```python
#Correlation between budget and gross
plt.figure(figsize=(10,6))
plt.scatter(df['budget'],df['gross'])
plt.title("Budget vs Gross scatter plot")
plt.show()
```

```python
#Budget vs Gross using regplot
plt.figure(figsize=(10,6))
sns.regplot(df['budget'],df['gross'],data=df,scatter_kws={"color":"red"},line_kws={"color":"green"})
plt.show()
```

```python
df.corr()
```

```python
cormax=df.corr()
plt.figure(figsize=(10,6))
sns.heatmap(cormax,annot=True,cmap="Greens")
plt.title("Correlation Matrix across numerical features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
```

```python
#Encoding Categorical Data

df_converted=df

for i in df_converted.columns:
    if(df_converted[i].dtype=="object"):
        df_converted[i]=df_converted[i].astype("category")
        df_converted[i]=df_converted[i].cat.codes
```

```python
df_converted
```

```python
#New correlation

cormax2=df_converted.corr()
plt.figure(figsize=(12,7))
sns.heatmap(cormax2,annot=True,cmap="Greens")
plt.title("Correlation Matrix across all features")
plt.xlabel("Movie Features")
plt.ylabel("Movie Features")
```

```python
corpairs=df_converted.corr()
corpairs=corpairs.unstack()
sorted_corpairs=corpairs.sort_values()
sorted_corpairs
```

```python
highestcor=sorted_corpairs[(sorted_corpairs)>0.5]
highestcor
```

# Hence we deduce that (votes,gross),(budget,gross) have the highest correlations. So movies with higher votes and budget tend to have higher gross amount
